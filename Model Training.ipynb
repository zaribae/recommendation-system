```python
df.iloc[7921]
```

```python
df['text'] = df['text'].str.lower().replace(r'^\w\s', ' ').replace(r'^\r\n', ' ', regex = True)
```

```python
print(df['text'][7921])
```

```python
import nltk
# nltk.download('punkt_tab')
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    # print(tokens)
    stemming = [stemmer.stem(w) for w in tokens]
    # print(stemming)
    return " ".join(stemming)
```

```python
df['text'] = df['text'].apply(lambda x: tokenization(x))
```

```python
print(df['text'][4])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['text'])
similarity = cosine_similarity(matrix)
  
```

```python
df_count_matrix = pd.DataFrame.sparse.from_spmatrix(matrix, columns=tfidvector.get_feature_names_out())
df_count_similarity = pd.DataFrame(similarity)
df_count_matrix.to_excel("output_tfidf.xlsx", engine='openpyxl')
df_count_similarity.to_excel("output_similarity.xlsx", engine='openpyxl')
```

```python
def recommendation(song_df):
    idx = df[df['song'] == song_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for m_id, sim_score in distances[1:21]:
        if sim_score > 0.4:
            songs.append(df.iloc[m_id].song)
            print(distances)

    return songs
```
